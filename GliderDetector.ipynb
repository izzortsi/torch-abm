```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

import torch
import torch.nn as nn
import torch.nn.functional as F

from IPython import display
import time
from math import *

from PIL import Image

from torchvision.transforms.functional import affine
```

```python
class ConvBlock(nn.Module):
    def __init__(self, NI, NO):
        super().__init__()
        
        self.l1 = nn.Conv2d(NI, NO, 3, padding=1, padding_mode='circular')
        self.l2 = nn.Conv2d(NO, NO, 3, padding=1, padding_mode='circular')
        self.bn = nn.BatchNorm2d(NO)
        
    def forward(self, x):
        z = F.elu(self.l1(x))
        z = self.bn(F.elu(self.l2(z)))
        return F.max_pool2d(z, 2)

class Attention(nn.Module):
    def __init__(self, KL):
        super().__init__()
        
        self.KL = KL
    
    def forward(self, k, q, v = None): # BS x KEYLEN x NK, BS x KEYLEN x NQ, BS x VALUE x NK
        w = F.log_softmax(torch.mean(k.unsqueeze(3) * q.unsqueeze(2), 1)*sqrt(self.KL), dim = 1) # BS x NK x NQ
        if v is None:
            y = 0
        else:
            y = torch.sum(v.unsqueeze(3) * torch.exp(w).unsqueeze(1), 2) # BS x VALUE x NQ
        
        return y, w

class GliderDetector(nn.Module):
    def __init__(self):
        super().__init__()
        
        self.c1 = ConvBlock(3, 32)
        self.c2 = ConvBlock(32, 64)
        self.c3 = ConvBlock(64, 128)
        self.ip1 = nn.Conv2d(128, 128, 3, padding=1, padding_mode='circular')
        self.ip2 = nn.Conv2d(128, 128, 3, padding=1, padding_mode='circular')
        self.c4 = ConvBlock(128, 256)
        
        self.attn = Attention(128)
        
        self.optim = torch.optim.Adam(self.parameters(), lr=1e-4)
        
    def forward(self, x, y):
        BS = x.shape[0]
        z = torch.cat([x,y],0)
        z = self.c1(z)
        z = self.c2(z)
        z = self.c3(z)
        z = F.elu(self.ip1(z))
        z = F.elu(self.ip2(z))
        z = self.c4(z)
        
        z = z.view(z.shape[0], z.shape[1], z.shape[2]*z.shape[3])
        
        z1 = z[:BS]
        z2 = z[BS:]
        
        y,w = self.attn(z1,z2,None)
        
        return w
```

```python
glider = GliderDetector().cuda()
#glider.load_state_dict(torch.load("glider_detector.pth"))
```

```python
TRAIN = 12000
TEST = 4000
BS = 10
RES = 512
FACTOR = 16

xx,yy = np.meshgrid(np.arange(RES), np.arange(RES))
zz = torch.cuda.FloatTensor(np.concatenate([xx[np.newaxis,np.newaxis,:,:], yy[np.newaxis,np.newaxis,:,:], np.ones((1,1,RES,RES))], axis=1))

def assembleBatch(i):
    x1 = []
    x2 = []
    lbl = []
    weight = []
    
    for j in range(BS):
        im = (np.array(Image.open("/sata/frames/gliders1/%.8d.png" % (i*BS+j+1)))-128.0)/128.0
        im = im[::4,::4,:].transpose(2,0,1)
        im = torch.cuda.FloatTensor(im).unsqueeze(0)

        x1.append(im)

        angle = np.random.rand()*360-180
        offset = [int( (np.random.rand()-0.5)*im.shape[1]), int( (np.random.rand()-0.5)*im.shape[2])]
        im2 = affine(im, angle=angle, translate=offset, scale=1, shear=[0,0])
        
        x2.append(im2)
        
        coord = affine(zz, angle=angle, translate=offset, scale=1, shear=[0,0])
        c = (coord/16).long()[:,:2,::FACTOR,::FACTOR].contiguous().view(1,2,(RES//FACTOR)**2)
        
        # Watch this line!
        c = c[:,0,:] + c[:,1,:]*(RES//FACTOR)
            
        lbl.append(c.view((RES//FACTOR)**2,))
        weight.append(coord[:,2,::FACTOR,::FACTOR].contiguous().view((RES//FACTOR)**2,))

    return torch.cat(x1,0), torch.cat(x2,0), torch.cat(lbl,0), torch.cat(weight,0)

tr_err = []
ts_err = []
idx = torch.arange(BS*((RES//FACTOR)**2)).long().cuda()

for epoch in range(10000):
    glider.train()
    errs = []
    for i in range(TRAIN//BS):
        x1,x2,lbl,wgt = assembleBatch(i)
        
        glider.optim.zero_grad()
        
        w = glider.forward(x1,x2)
        w = w.permute(0,2,1).contiguous().view(BS*(RES//FACTOR)**2, (RES//FACTOR)**2)
        
        loss = -torch.sum(wgt*w[idx,lbl[idx]])/torch.sum(wgt)
        loss.backward()
        
        glider.optim.step()
        errs.append(loss.cpu().detach().item())
        del loss
        
    tr_err.append(np.mean(errs))
    
    glider.eval()
    with torch.no_grad():
        errs = []
        for i in range(TEST//BS):
            x1,x2,lbl,wgt = assembleBatch(i+TRAIN//BS)

            w = glider.forward(x1,x2) # BS x (RES/16)**2 x (RES/16)**2
            w = w.permute(0,2,1).contiguous().view(BS*(RES//FACTOR)**2, (RES//FACTOR)**2)

            loss = -torch.sum(wgt*w[idx,lbl[idx]])/torch.sum(wgt)

            errs.append(loss.cpu().detach().item())
        del loss
    ts_err.append(np.mean(errs))
      
    w = torch.exp(w).cpu().detach().numpy().reshape((BS, RES//FACTOR, RES//FACTOR, RES//FACTOR, RES//FACTOR))
    plt.clf()
    plt.subplot(2,2,1)
    plt.plot(tr_err)
    plt.plot(ts_err)
    
    plt.subplot(2,2,2)    
    plt.imshow(w[0,8,8])

    plt.subplot(2,2,3)    
    plt.imshow(w[0,12,8])

    plt.subplot(2,2,4)    
    plt.imshow(w[0,8,12])
    
    display.clear_output(wait=True)
    display.display(plt.gcf())
    time.sleep(0.01)
```

```python
def extractPairs(w): # Doesn't take a whole batch...
    XR = w.shape[0]
    YR = w.shape[1]
    ww = w.reshape(XR*YR, XR*YR)
    
    best = np.argmax(ww, axis=0)
    bestval = ww[best[np.arange(XR*YR)], np.arange(XR*YR)]

    sink = np.where(bestval>0.2)[0]
    src = best[sink]
    
    src_x = src%XR
    src_y = src//XR
    
    sink_x = sink%XR
    sink_y = sink//XR
    
    return np.concatenate([src_x[:,np.newaxis], src_y[:,np.newaxis]], axis=1), np.concatenate([sink_x[:,np.newaxis], sink_y[:,np.newaxis]], axis=1)
```
